```python
#NEW PROBLEM
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#fig = plt.figure()
#ax = fig.add_subplot(1, 1, 1)

titanic=pd.read_csv('/Users/klombardo/Garbage/titanic.csv', usecols=['Sex','Age','Survived']) #,index_col='Name') #, names=['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'])
titanic=titanic.dropna(subset = ['Age']).sort_values(by=['Age'])

df = pd.DataFrame({"Females Survived":titanic[titanic['Sex'] == 'female'].groupby('Age')['Survived'].sum(),"Males Survived":titanic[titanic['Sex'] == 'male'].groupby('Age')['Survived'].sum()},index=titanic['Age'])
ax = df.plot.bar(color=["Blue","Red"], rot=0, title="Survivors")

```
